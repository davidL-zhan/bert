```python
import sys
from pathlib import Path

cwd = Path.cwd()
project_root = cwd.parent if cwd.name == "demo" else cwd

if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))
from transformers import BertTokenizer
from config import config
tokenizer = BertTokenizer.from_pretrained(config.model_name)
```

```python
senteces = [
    "对现在的化妆品真是没有抵抗能力，店家宣传、别人一说好就晕了头跟风去买。其实最简单的东西也是最好的。看了这本书，更注意从内而外调养。",
    "比较不错的酒店，设施比较新，交通还是比较方便，但是离市中心有一点距离，去机场方便。",
]
encoded = tokenizer(
    senteces,
    padding=True,
    truncation=True,
    max_length=config.max_length,
    return_tensors="pt",
)
print(encoded)
```
